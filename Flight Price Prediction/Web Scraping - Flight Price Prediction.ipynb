# Flight Price Prediction - Web Scraping

```python
# importing libraries/dependencies
import re
import time
import requests
import warnings
import selenium
import numpy as np
import pandas as pd
from selenium import webdriver
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import ElementNotVisibleException
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, ElementClickInterceptedException
```

```python
# Running webdriver
driver = webdriver.Chrome(r'C:\FLIP ROBO\Crome Driver\chromedriver.exe')
driver.maximize_window()
```

```python
# getting the webpage of mentioned url
url = "https://www.makemytrip.com/"
driver.get(url)
time.sleep(3)
```

```python
# Closing notification
driver.find_element (By.XPATH, '//*[@id="SW"]/div[1]/div[1]/ul/li[4]').click()
```

```python
# Clicking on search
driver.find_element(By.XPATH, '//*[@id="root"]/div/div[2]/div/div/div[2]/p/a').click()
time.sleep(5)
```

```python
# Closing notification
driver.find_element (By.XPATH, '//*[@id="root"]/div/div[2]/div[2]/div[2]/div/span').click()
```

```python
# Creating empty lists

Flight_Name= [] # For flight company/name
Source = [] # Starting point
Destination= [] # End/Destination point
Duration = [] # Time taken to reach the destination
Stops = [] # If any stops between source to destinations
Price = []# Price of the flight
Arrival = []# Arrival time
Departure = []# Departure time
```

```python
# Giving variable to "dirver.find_elements", further it will be easy just to add X instead of full code.
X = driver.find_elements
```

```python
# Creating list of few busy destinations in india
loc_lst = ['New Delhi', 'Mumbai', 'Bengaluru', 'Hyderabad', 'Kolkata', 'Chennai', 'Ahmedabad', 'Dabolim','Kochi', 'Pune']
```

```python
# Extracting information of flights using for loop and 'try and except'
for x in loc_lst: #Running location list for from (from_)
    for y in loc_lst:# Running location list to destination (to_)
        if x!=y:# x is not equal to y as source and destination can not be the same
            from_ = driver.find_element(By.XPATH, '/html/body/div[1]/div/div[2]/header/div[1]/div/div[2]').click()
            a = driver.find_element(By.XPATH , "/html/body/div[1]/div/div[2]/header/div[1]/div/div[2]/div[1]/form/div/input")
            a.send_keys(x)
            time.sleep(2)
            driver.find_element(By.XPATH, '/html/body/div[1]/div/div[2]/header/div[1]/div/div[2]/div[1]/form/div/div/div/ul/li[1]').click()
            time.sleep(2)
            
            to_ = driver.find_element(By.XPATH, '//*[@id="search-widget"]/div/div[4]').click()
            a = driver.find_element(By.XPATH , "/html/body/div[1]/div/div[2]/header/div[1]/div/div[4]/div[1]/form/div/input")
            a.send_keys(y)
            time.sleep(2)
            driver.find_element(By.XPATH, '/html/body/div[1]/div/div[2]/header/div[1]/div/div[4]/div[1]/form/div/div/div/ul/li[1]').click()
            driver.find_element (By.XPATH, '//*[@id="search-widget"]/div/div[5]/div[1]/div/div/div/div/div[2]/div[1]/div[3]/div[2]/div[2]').click()
            
            
            # Clicking on search button
            search = driver.find_element(By.XPATH, '//*[@id="search-button"]').click()
            time.sleep(10)
            
                    
            # Closing 1st notification 
            try:
                driver.find_element (By.XPATH,'/html/body/div[1]/div/div[2]/div[2]/div[2]/div/span').click()
            except NoSuchElementException:
                pass
            
            # Closing second Notification
            try:
                driver.find_element(By.XPATH, '/html/body/div[1]/div/div[2]/div[2]/div[2]/div/div/div[3]/button').click()
            except NoSuchElementException:
                pass
            
            # Extracting Flight Names
            try:
                for i in X(By.XPATH, '//p[@class="boldFont blackText airlineName"]'):
                    Flight_Name.append(i.text)
            except NoSuchElementException:
                pass
                        
            # Extracting Source Names
            try:
                source = X(By.XPATH,'//p[@class="blackText"]')   
                for i in source[::2]:
                    Source.append(i.text)
            except NoSuchElementException:
                pass
            
            # Extracting Destination name
            try:
                destination = X (By.XPATH, '//p[@class="blackText"]')
                for i in destination[1::2]:
                    Destination.append(i.text)
            except NoSuchElementException:
                pass
            
            # Extracting Duration time
            try:
                duration = X (By.XPATH, '//div[@class="stop-info flexOne"]//p')
                for i in duration[::3]:
                    Duration.append(i.text)
            except NoSuchElementException:
                pass
            
            # Extracting stops
            try:
                for i in X (By.XPATH, '//p[@class="flightsLayoverInfo"]'):
                    Stops.append(i.text.split('via')[0])
            except NoSuchElementException:
                pass
            
            # Extracting Price
            try:
                for i in X (By.XPATH, '//p[@class="blackText fontSize18 blackFont white-space-no-wrap"]'):
                    Price.append(i.text)
            except NoSuchElementException:
                pass
            
            # Extracting Arrival
            try:
                arrival = X (By.XPATH, '//p[@class="appendBottom2 flightTimeInfo"]')
                for i in arrival[1::2]:
                    Arrival.append(i.text.replace('\n',''))
            except NoSuchElementException:
                pass
            
            # Extracting Departure
            try:
                departure = X (By.XPATH, '//p[@class="appendBottom2 flightTimeInfo"]')
                for i in departure[::2]:
                    Departure.append(i.text)
            except NoSuchElementException:
                pass
            
            
```

```python
# Creating column name of data frame
df = pd.DataFrame({'Flight_Name':Flight_Name,'Source':Source,'Destination':Destination,'Duration':Duration,
              'Stops':Stops, 'Departure':Departure, 'Arrival':Arrival ,'Price':Price})
```

```python
# Remvoing extra information from Arrival column
df['Arrival'] =  df['Arrival'].apply(lambda x: x.replace('+ 1 DAY', ''))
```

```python
# Replacing rupee symbol with blank space
df['Price'] =  df['Price'].apply(lambda x: x.replace('₹', ''))
```
